```python
from datasets import load_dataset

ds = load_dataset("antoniomenezes/go_emotions_ptbr")
```

```python
ds
```

```python
df = ds["train"].to_pandas()

```

```python

ids = df["id"].value_counts()

# Get ids with more than 3 ratings
ids = set(ids[ids >= 3].index)
```

```python
ds = ds.filter(lambda x: x["id"] in ids)
```

```python
df = ds["train"].to_pandas()
```

```python
df.shape
```

```python
columns = [ 
    'id',
    'example_very_unclear',
    'admiration',
    'amusement',
    'anger',
    'annoyance',
    'approval',
    'caring',
    'confusion',
    'curiosity',
    'desire',
    'disappointment',
    'disapproval',
    'disgust',
    'embarrassment',
    'excitement',
    'fear',
    'gratitude',
    'grief',
    'joy',
    'love',
    'nervousness',
    'optimism',
    'pride',
    'realization',
    'relief',
    'remorse',
    'sadness',
    'surprise',
    'neutral',
]

df_agg = df[columns].groupby("id")
```

```python

# Add a count column
df_agg = df_agg.sum()


df_agg
```

```python
df_agg = df_agg[df_agg["example_very_unclear"] == 0]

df_agg
```

```python
emotions = columns[2:]

df_final = df_agg[emotions] >= 2
```

```python
df_final.sum()
```

```python
# Add the text columns!

df_final = df_final.join(df[["id", "text", "texto"]].set_index("id"))
```

```python
df_final.columns
```

```python
df_final.rename(columns={"texto": "text", "text": "text_original"}, inplace=True)

```

```python
# split train dev test

from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df_final, test_size=0.3, random_state=42)

df_train, df_dev = train_test_split(df_train, test_size=0.2, random_state=42)

df_train.shape, df_dev.shape, df_test.shape
```
